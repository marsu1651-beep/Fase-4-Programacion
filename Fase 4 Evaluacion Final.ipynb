## 7. Scatter Plot

```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.DataFrame({
    "gdp": np.random.uniform(1000, 70000, 50),
    "co2": np.random.uniform(1, 25, 50),
    "population": np.random.uniform(1000000, 1000000000, 50),
    "continent": np.random.choice(
        ["Asia", "Europe", "America"],
        50
    )
})

fig = px.scatter(
    df,
    x="gdp",
    y="co2",
    size="population",
    color="continent",
    title="Relación PIB vs CO2"
)

fig.show()
```
